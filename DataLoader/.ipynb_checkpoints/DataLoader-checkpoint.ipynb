```python
import torch
import torchvision
import torch.nn as nn
```

```python
training_data=torchvision.datasets.MNIST("./",transform=torchvision.transforms.ToTensor(),download=True,train=True)
```

```python
testing_data=torchvision.datasets.MNIST("./",transform=torchvision.transforms.ToTensor(),download=True,train=False)
```

```python
batch_size=256
train_loader=torch.utils.data.DataLoader(dataset=training_data,shuffle=True,batch_size=batch_size)
test_loader=torch.utils.data.DataLoader(dataset=testing_data,shuffle=True,batch_size=batch_size)
```

```python
#for batch_number ,(images,labels) in enumerate(train_loader):
#    print(images.shape)
device= "cuda" if torch.cuda.is_available else "cpu"
print(device)
```

```python
"""
nn.Sequential(nn.Conv2d(1,32,3),
              nn.ReLU(),
              nn.MaxPool2d(3,2),
              nn.Conv2d(32,64,3),
              nn.ReLU(),
              nn.MaxPool2d(64,2),
              nn.Linear(64*7*7,1024),
              nn.Linear(1024,10))
              """
```

```python
num_features=10
class Net(nn.Module):
    def __init__(self,num_features):
        super(Net,self).__init__()
        #Layer one
        self.conv1=nn.Sequential(nn.Conv2d(1,32,3),
                                nn.ReLU(),
                                nn.MaxPool2d(3,2))
        #Layer tow
        self.covn2d_2=nn.Conv2d(32,64,(3,3))
        self.relu_2=nn.ReLU()
        self.max2d_2=nn.MaxPool2d(3,2)
        #Layer three
        self.fc_1=nn.Linear(64*7*7,1024)
        self.fc_2=nn.Linear(1024,num_features)
    def forward(self,x):
        #Layer1
        y=self.covn1(x)
        #Layer2
        y=self.covn2d_2(y)
        y=self.relu_2(y)
        y=self.max2d_2(y)
        #Layer3
        y=y.view(y.size(0),-1)#N-bachsize*3136
        y=self.fc_1(y)
        y=self.fc_2(y)
        return y
```

```python
model=Net(num_features)
```

```python
from torch import optim
criterion=nn.CrossEntropyLoss()
optimizer=optim.SGD(model.parameters(),lr=0.001,momentum=0.9)
```

```python
import time
import copy
def train_model(model,dataloaders,criterion,optimizer,num_epochs=25):
    scince=time.time
    val_acc_history=[] #best Weights are here
    best_model_wts=copy.deepcopy(model.state_dict()) # copy a weight
    best_acc=0.0
    for epoch in range(num_epochs):
        print("{}/{}".format(epoch,num_epochs-1))
        print("-"*10)
        
        # each epoch has a train & validation phase
        
        for phase in ["train","val"]:
            if phase == "train":
                model.train() # set model to train model
            else:
                model.eval() # set model to evaluate model
        
            running_loss=0.0
            running_currects=0
            for inputs,labels in dataloaders[phase]:
                inputs=inputs.to(device)
                labels=labels.to(device)
                optimizer.zero_grad()
                
                with torch.set_grad_enabled(phase=="train"):
                    outputs=model(inputs)
                    loss=criterion(outputs,labels)
                    _,preds=torch.max(outputs,1)
                    if phase =="train":
                        loss.backward()
                        optimizer.step()
                        
                    running_loss+=loss.item()*inputs.size(0)
                    running_currects+=torch.sum(preds,labels.data)
                    
                epoch_loss=running_loss/len(dataloaders[phase].dataset)
                epoch_acc=running_currects.double()/len(dataloaders[phase].dataset)
                print("{}loss: {:.4f} acc: {:.4f}".format(phase,epoch_loss,epoch_acc))
                
                if phase=="val" and epoch_acc>best_acc:
                    best_acc=epoch_acc
                    best_model_wts = copy.deepcopy(model.state_dict())
                if phase=="val":
                    val_acc_history.append(epoch_acc)
            print()
            
            time_elapsed=time.time()-scince
            print("training complete in {:.0f}m {:.0f}s".format(time_elapsed//60,time_elapsed%60))
            print("Best Val Acc {:.4f}".format(best_acc))
            model.load_state_dict(best_model_wts)
            return model,val_acc_history
        
                    
                    
                    
                    
                    
```

```python
train_loader=dict(train_loader)
test_loader=dict(test_loader)
from collections import Container
datas=dict(zip(train_loader,test_loader))
```

```python
model_fit,hist=train_model(model,datas,criterion,optimizer,num_epochs=25)
```
