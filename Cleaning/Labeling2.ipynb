```python
import pandas as pd
```

```python
original_dataset=pd.read_csv("Electric_Production.csv")
original_dataset
```

```python
from sklearn.preprocessing import scale,normalize,minmax_scale
```

```python
app=""
app_2=""
x=original_dataset.iloc[:,0]
for date in x:
    splited=str(date.split(","))
    app+=splited
for date in app:
    splited=date.strip("[/]")
    app_2+=splited
x=app_2.split("''")
```

```python
new_cols=[]
for q in x:
    try:
        new_cols.append(int(q))
    except:
        new_cols.append(111999)
```

```python
original_dataset.drop("DATE",axis=1,inplace=True)
original_dataset
```

```python
Q=pd.DataFrame(new_cols)
Q.rename(columns={0:"DATE"},inplace=True)
```

```python
new_dataset=pd.concat([Q,original_dataset],ignore_index=True,axis=1)
```

```python
new_dataset
```

```python
new_dataset.rename(columns={0:"DATE",1:"IPG"},inplace=True)
```
