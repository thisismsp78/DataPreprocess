```python
import pandas as pd
```

```python
original_dataset=pd.read_csv("sales-of-shampoo-over-a-three-ye.csv")
original_dataset
```

```python
def month_string_to_number(string):
    m = {
        'jan': 1,
        'feb': 2,
        'mar': 3,
        'apr':4,
         'may':5,
         'jun':6,
         'jul':7,
         'aug':8,
         'sep':9,
         'oct':10,
         'nov':11,
         'dec':12
        }
    s = string.strip()[:3].lower()

    try:
        out = m[s]
        return out
    except:
        raise ValueError('')
```

```python
lists=[]
x=original_dataset.iloc[:,0]
for q in x:
    lists.append(q.strip("123-"))
```

```python
new_cols=[]
for month in lists:
    new_cols.append(month_string_to_number(month))
```

```python
new_cols
```

```python
days=[]
for x in range(1,4):
    for y in range(1,13):
        days.append(x)
```

```python
days=pd.DataFrame(days)
month=pd.DataFrame(new_cols)
```

```python
new_dataset=pd.concat([month,days,original_dataset["Sales of shampoo over a three year period"]],ignore_index=False,axis=1)
```

```python
new_dataset
```

```python
new_dataset=pd.DataFrame(new_dataset.values,index=list(range(36)),columns=["Month","Days","Sales"])
```
