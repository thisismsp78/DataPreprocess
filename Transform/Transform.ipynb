```python
import numpy as np
file=np.array(csv_file.values)
file
```

```python
from sklearn.preprocessing import Imputer
imp=Imputer(missing_values="NaN",strategy="mean",axis=0)
imp.fit(file[:,2:])
file[:,2:]=imp.transform(file[:,2:])
```

```python
file
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
ct=ColumnTransformer(transformers=[("encoder",OrdinalEncoder(),[0,1])],remainder="passthrough")
file_end=np.array(ct.fit_transform(file))
```

```python
print(file_end)
```

```python
new=pd.DataFrame(file_end,index=csv_file.index,columns=csv_file.columns)
```
